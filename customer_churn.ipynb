```python
import pandas as pd
```

```python
file_name = ''
df = pd.read_csv(file_name)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', 0, regex=False)
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
df.replace({'Yes':1, 'No':0}, inplace=True)
df.replace({'Male':1, 'Female':0}, inplace=True)
df.replace({'No phone service':2},inplace=True)
df['InternetService'].replace({'DSL':1, 'Fiber optic':2}, inplace=True)
df.replace({'No internet service':2},inplace=True)
df['Contract'].replace({'Month-to-month':0, 'One year':1, 'Two year':2}, inplace=True)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
df.columns
```

```python
x = df[['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'MonthlyCharges', 'TotalCharges']]
y = df['Churn']
#churn(y) is our output, and all the other columns that has effect on the output are given as input(x)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
#using LOGISTIC REGRESSION model
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state=2529, train_size=0.75)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
#give your own input to test the model for its accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
#81% accuracy achieved with 75% train test split
```
